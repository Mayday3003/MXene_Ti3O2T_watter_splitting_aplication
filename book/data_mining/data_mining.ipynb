# Matminer data mining

```python
from matminer.datasets import get_available_datasets, load_dataset
import io, contextlib, warnings

warnings.filterwarnings("ignore")  # opcional: silencia warnings

CANDIDATE_COLS = [
    "formula", "formula_pretty", "pretty_formula",
    "chemical_formula", "composition"
]
TARGET = ["Ti", "C","O"]

# obtener lista de datasets sin imprimir la descripción
buf = io.StringIO()
with contextlib.redirect_stdout(buf):
    datasets = get_available_datasets()

if isinstance(datasets, dict):
    dataset_names = list(datasets.keys())
else:
    dataset_names = [d if isinstance(d, str) else d[0] for d in datasets]

for ds_name in dataset_names:
    try:
        # cargar dataset sin que imprima mensajes
        buf = io.StringIO()
        with contextlib.redirect_stdout(buf):
            df = load_dataset(ds_name, download_if_missing=False)
    except Exception:
        continue

    found_col = next((c for c in CANDIDATE_COLS if c in df.columns), None)
    if not found_col:
        continue

    for idx, val in df[found_col].dropna().items():
        val_str = str(val)
        if all(el in val_str for el in TARGET):
            print(f"{ds_name}\tindex={idx}\tcol={found_col}\tformula={val_str}")
```

# Figshare

**Figshare.** Figshare provides open research-repository infrastructure that helps researchers and institutions share, showcase, and manage research outputs in a discoverable, citable, and transparent manner. ([About Figshare](https://info.figshare.com/about/))

Users can create accounts to download a wide range of scientific materials and to publish supplementary data associated with their work.

On Figshare we located a particularly relevant dataset uploaded in 2024 by Valentina Nesterova, Vladislav Korostelev, and Konstantin Klyukin: a compressed archive containing crystallographic information files (.cif) associated with their article *“Unveiling the Role of Termination Groups in Stabilizing MXenes in Contact with Water.”* The archive includes eight .cif files representing $Ti_{3}C_{2}T_{2}$ MXene surfaces with different termination groups. ([Figshare dataset](https://figshare.com/articles/dataset/Ti3C2T2_zip/25236772?utm_source=chatgpt.com&file=44583274))

This dataset is especially important because many publications refer to the material generically as $Ti_{3}C_{2}T_{x}$, and recent experimental or computational details about specific termination groups may not yet be incorporated into broader resources such as the Materials Project or Matminer.
