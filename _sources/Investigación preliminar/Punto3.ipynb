# Búsqueda bibliográfica en Scopus sobre MXenes

```python
#Liberías a utlizar 
import pandas as pd 
import matplotlib.pyplot as plt
from pathlib import Path
```

### Creación dataframes

```python
carpeta = Path("busqueda")
dfs = {}  

for ruta in carpeta.glob("*.csv"):
    nombre = ruta.stem               
    dfs[nombre] = pd.read_csv(ruta)

print(dfs.keys())
```

```python
df_mxenes_xrd_catalysis = dfs["mxenes_xrd_catalysis"]  #Dataframe con busqueda MXenes AND XRD AND catalysis

df_mxenes = dfs["mxenes"]  #Dataframe con busqueda MXenes

df_mxenes_xrd_catalysis_bandgap = dfs["mxenes_xrd_catalysis_bandgap"]      

df_mxenes_xrd = dfs["mxenes_xrd"]  #Dataframe con busqueda MXenes AND XRD

df_mxenes_xrd_bandgap = dfs["mxenes_xrd_bandgap"] #Dataframe con busqueda MXenes AND XRD AND catalysis

df_mxenes_catalysis = dfs["mxenes_catalysis"] #Dataframe con busqueda MXenes AND catalysis

df_mxenes_bandgap_catalysis = dfs["mxenes_bandgap_catalysis"] #Dataframe con busqueda Mxenes AND bandgap AND catalysis

df_mxenes_bandgap_catalysis = dfs["mxenes_bandgap_catalysis"] #Dataframe con busqueda Mxenes AND bandgap AND catalysis
```

# Gráfica general

```python
plt.figure(figsize=(10,5))

df_mxenes_water = pd.read_csv('busqueda/mxenes_water_s.csv')
df_mxenes_water.columns = ["Year", "Count"]



plt.figure(figsize=(10,5))
plt.plot(df_mxenes_water["Year"], df_mxenes_water["Count"], marker="o", linestyle="-"
         , label="MXenes + water splitting")



plt.plot(df_mxenes_xrd_catalysis["Year"], df_mxenes_xrd_catalysis["Count"],
         marker="o", linestyle="-", label="MXenes + XRD + Catalysis")

plt.plot(df_mxenes["Year"], df_mxenes["Count"],
         marker="o", linestyle="-", label="MXenes (general)")

plt.plot(df_mxenes_xrd_catalysis_bandgap["Year"], df_mxenes_xrd_catalysis_bandgap["Count"],
         marker="o", linestyle="-", label="MXenes + XRD + Catalysis + Bandgap")

#plt.plot(df_mxenes_xrd["Year"], df_mxenes_xrd["Count"],
#         marker="o", linestyle="-", label="MXenes + XRD")

plt.plot(df_mxenes_xrd_bandgap["Year"], df_mxenes_xrd_bandgap["Count"],
         marker="o", linestyle="-", label="MXenes + XRD + Bandgap")

plt.plot(df_mxenes_catalysis["Year"], df_mxenes_catalysis["Count"],
         marker="o", linestyle="-", label="MXenes + Catalysis")

plt.legend()
plt.title("Artículos por año")
plt.xlabel("Año")
plt.ylabel("Número de artículos")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)  # opcional: rota las etiquetas del eje x para mejor lectura
plt.show()
```

# Análisis de tendencias

Las gráficas presentadas permiten observar la evolución de la investigación sobre MXenes en diferentes contextos y aplicaciones a lo largo del tiempo.

**1. Gráfica general de todas las búsquedas:**
La gráfica general muestra un crecimiento sostenido en el número de publicaciones científicas relacionadas con MXenes y sus aplicaciones (XRD, catalysis, bandgap, etc.). Se observa que,en la última década, la investigación sobre los MXenes ha aumentado considerablemente, reflejando su relevancia en la ciencia de materiales y sus múltiples aplicaciones potenciales. Las búsquedas combinadas (por ejemplo, MXenes + XRD + Catalysis) presentan un menor número de artículos, lo que indica que la investigación en áreas multidisciplinarias aún está en desarrollo, pero con una tendencia creciente.

**2. Gráfica de MXenes en water splitting:**
La gráfica específica de "MXenes AND water splitting" muestra que el interés en esta aplicación es más reciente y el número de publicaciones es menor en comparación con las búsquedas generales. Sin embargo, se aprecia una tendencia ascendente en los últimos años, lo que sugiere que el uso de MXenes en procesos de separación de agua (water splitting) está emergiendo como un campo prometedor, probablemente impulsado por la búsqueda de materiales eficientes para la producción de hidrógeno y energías limpias.

**Conclusiones:**
- El crecimiento general en publicaciones sobre MXenes refleja su importancia y versatilidad en la investigación actual.
- Las aplicaciones específicas, como el water splitting, aunque más recientes y menos numerosas, muestran un aumento acelerado, lo que podría indicar un futuro prometedor en este campo.
- La tendencia sugiere que, a medida que se descubren nuevas propiedades y aplicaciones de los MXenes, surgirán más investigaciones en áreas especializadas, como la catálisis y la producción de hidrógeno.
