### $H_{2}$ como combustible renovable


#### 2. Por qu el hidrgeno?


- **Cero emisiones directas en uso**: cuando se usa como combustible su emisin directa es agua

- **Alta densidad energtica por masa**: El hidrgeno tiene la mayor energa por kilogramo de cualquier combustible conocido $(~120 MJ/kg)$, casi 3 veces ms que la gasolina.
Esto lo hace ideal para aplicaciones donde el peso importa: aviacin, transporte pesado y espacial.


#### Desafos actuales

- Alto costo de produccin.

- Dificultad de almacenamiento y transporte (es muy liviano y requiere alta presin o temperaturas criognicas).

- Infraestructura limitada (pocos gasoductos y estaciones de recarga).

- Requiere mucha energa elctrica

## Qu son los MXenes?

Los *MXenes* son materiales bidimensionales de la familia de los carburos, nitruros y carbonitruros de metales de transicin con frmula general $M_{n+1}X_nT_x$, donde $M$ = metal de transicin, $X$ = C y/o N, $T_x$ representa grupos de terminacin de superficie (p. ej. O, OH, F).
